```
# Sigmoid Function or Logistic Function Characteristic
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-6,+6,200)
y = 1 / (1 + np.exp(-x))

plt.plot(20+x, y)
plt.xlabel('size of tumour')
plt.ylabel('malignant or benign')
plt.show()
```
